# Assignment

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('house_prices.csv')
df.info()
```

# 1.&nbsp; Count of one categorical feature

```python
fig, ax = plt.subplots()
unique_conditions = df['SaleCondition'].unique()

# Create bar chart with centered bars
ax.bar(range(len(unique_conditions)), df['SaleCondition'].value_counts())


ax.set_xticks(range(len(df['SaleCondition'].unique())))
ax.set_xticklabels(df['SaleCondition'].unique(), rotation=90)

# Set the title
ax.set_ylabel("Number of House")
ax.set_xlabel("Sale Condition")
ax.set_title("Bar Chart of House Sale Conditions")

# Add labels on top of the bars
for i, value in enumerate(df['SaleCondition'].value_counts()):
    ax.text(i, value + 0.1, str(value), ha='center', va='bottom')
```

Berdasarkan visualisasi diperoleh informasi mengenai Distribusi Kondisi Penjualan:
   - Kondisi penjualan paling tinggi adalah 'Normal', yang menunjukkan bahwa sebagian besar penjualan berlangsung dalam kondisi penjualan normal.
   - Kondisi penjualan 'Family' memiliki frekuensi yang paling rendah, menunjukkan bahwa penjualan dalam konteks keluarga atau hubungan keluarga mungkin kurang umum.
   - Kondisi penjualan lainnya seperti 'Abnormal', 'Partial','AdjLand', dan 'Alloca' lebih jarang terjadi, dan bisa jadi memiliki keadaan penjualan yang lebih spesifik atau tidak biasa.

# 2.&nbsp; Distribution of one continuous feature

```python
fig, ax = plt.subplots()
sns.histplot(df['SalePrice'], ax=ax, kde=True, bins=50)

#title
ax.set_ylabel("Number of House")
ax.set_xlabel("Sale Price")
ax.set_title("Histogram of House Sale Price")
```

Berdasarkan visualisasi diperoleh informasi mengenai Distribusi Harga Penjualan rumah:
   - distribusi harga jual rumah terlihat mirip dengan distribusi normal
   - Skewness ke kiri menunjukkan bahwa sebagian besar rumah memiliki harga jual yang lebih rendah, sementara sejumlah kecil rumah memiliki harga jual yang lebih tinggi dari rata-rata.

# 3.&nbsp; Categorical-continuous

```python
sns.stripplot(
    data=df,
    y="SalePrice",
    x="Neighborhood",
    hue="SaleType",
    marker='o',
)
# Calculate the overall average sale price
overall_average_price = df["SalePrice"].mean()

# Draw a horizontal line for the overall average
plt.axhline(y=overall_average_price, color='black', linestyle='--', linewidth=2, label='Sale Price Average')

# Rotate x-axis labels by 90 degrees
plt.xticks(rotation=90)

# Set labels and title
plt.xlabel("Neighborhood")
plt.ylabel("Sale Price")
plt.title("Strip Plot of Sale Price vs. Neighborhood")

# Add legend outside the plot (assuming there's a legend)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

Berdasarkan visualisasi diperoleh informasi mengenai Distribusi Harga Penjualan rumah berdasarkan Neighborhood:
   - Terlihat variasi yang signifikan dalam harga penjualan rumah antar neighborhood.
   - Beberapa neighborhood memiliki range harga yang lebih luas dibanding yang lain,seperti: NoRidge, OldTown, NridgHt dan StoneBr
   - Beberapa neighborhood memiliki harga yang lebih terkonsentrasi, seperti: MeadowV, NPkVill dan BrDale
   - Beberapa neighborhood memiliki kecenderungan kisaran harga diatas rata-rata, seperti: Veenker, NoRidge, Somerst, NridgHt, Timber, Gilbert dan StoneBr
   Beberapa neighborhood memiliki kecenderungan kisaran harga dibawah rata-rata, seperti: IDOTRR, MeadowV, NPkVill, BrDale dan Blueste
   - SaleType 'New' memiliki kecenderungan harga jual lebih tinggi dibanding yang lain

# 4.&nbsp; Continuous-continuous

```python
sns.scatterplot(
    data=df, 
    x="YearBuilt", 
    y="SalePrice",
    hue="OverallQual", 
    palette="viridis",  
    alpha=0.8, 
    marker='o', 
    color='blue'  # You can customize the color as needed
)
# Calculate the overall average sale price
overall_average_price = df["SalePrice"].mean()

# Draw a horizontal line for the overall average
plt.axhline(y=overall_average_price, color='black', linestyle='--', linewidth=2, label='Sale Price Average')

# Set labels and title
plt.xlabel("Year Built")
plt.ylabel("Sale Price")
plt.title("Scatter Plot of Year Built vs. Sale Price")

# Add legend outside the plot (assuming there's a legend)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

Berdasarkan visualisasi diperoleh informasi mengenai hubungan Harga Penjualan rumah berdasarkan tahun pembangunan:
   - Terdapat kecenderungan bahwa harga penjualan cenderung naik seiring dengan tahun pembangunan. Rumah yang lebih baru atau lebih baru cenderung memiliki harga penjualan yang lebih tinggi dibandingkan dengan rumah yang lebih tua.
   - Rumah yang memiliki kualitas tinggi (terutama dengan nilai 10) cenderung memiliki harga diatas rata-rata
