```python
syarat = df['tenure'] != 0
df_baru = df[syarat]
df_baru
```

```python
df_baru.info()
```

```python
df_baru.loc[:, 'TotalCharges'] = pd.to_numeric(df_baru['TotalCharges'], errors='coerce').astype('float64')
df_baru.info()
```

```python
#rata rata tenure tidak memiliki partner dan dependent
syarat_1=df_baru['Partner']=='No'
syarat_2=df_baru['Dependents']=='No'

df_no2=df_baru[syarat_1 & syarat_2]
tenure_average = df_no2['tenure'].mean()

print("Rata-rata tenure yang tidak memiliki partner dan dependent:", tenure_average)
```

```python
#max monthly changer memiliki partner dan tidak memiliki dependent
syarat_1=df_baru['Partner']=='Yes'
syarat_2=df_baru['Dependents']=='No'

df_no3=df_baru[syarat_1 & syarat_2]
MonthlyCharges_max = df_no3['MonthlyCharges'].max()

print("Max MonthlyCharges yang memiliki partner dan tdk memiliki dependent:", MonthlyCharges_max)
```

```python
#jumlah costumer tidak memiliki partner dan memiliki dependent
syarat_1=df_baru['Partner']=='No'
syarat_2=df_baru['Dependents']=='Yes'

df_no4=df_baru[syarat_1 & syarat_2]
customerID_count = df_no4['customerID'].count()

print("Jumlah costumer yang tidak memiliki partner dan memiliki dependent:", customerID_count)
```

```python
#TOTAL CHARGES dari customer NON-SENIOR CITIZEN (Senior Citizen = 0) yang memiliki PHONE SERVICE dan memiliki INTERNET SERVICE FIBER OPTIC
syarat_1=df_baru['SeniorCitizen']==0
syarat_2=df_baru['PhoneService']=='Yes'
syarat_3=df_baru['InternetService']=='Fiber optic'

df_no5=df_baru[syarat_1 & syarat_2 & syarat_3]
TotalCharges_average = df_no5['TotalCharges'].mean()

print("Total Charge rata-rata dari non senior citizen yang memiliki phine service dan internet service fiber optic:", TotalCharges_average)
```
